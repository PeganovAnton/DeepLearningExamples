```python
import glob, os, sys
```

```python
files = glob.glob('final_tfrecords_sharded_gvocab_base/*tfrecord*')
```

```python
files
```

```python
len(files)
```

```python
prefix = 'final_tfrecords_sharded_gvocab_base/tf_examples.'
for filei in files:
    fname = filei.split('.')[-1]
    numi = int(fname[8:])
    
    if numi < 10:
        print(filei)
        continue
    elif numi >= 10 and numi < 100:
        suffix = '00' + str(numi)
    elif numi >= 100 and numi < 1000:
        suffix = '0' + str(numi)
    elif numi >= 1000:
        suffix = str(numi)
    else:
        print(fname)
        print('!!!')
        break
    
    fname_ = fname[:8] + suffix
    newfname =  prefix + fname_
    
    os.rename(filei, newfname)
```
